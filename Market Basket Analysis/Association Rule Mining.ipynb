```python
records = []
for i in range(rows):
  records.append([str(df.values[i,j]) for j in range(cols)])
records[:10]
```

**To build the association rules, we need to build the criteria**\

    support : based on data
    confidence : 0.2
    lift : 3

    support : transactions - weekly data: 7days , 5
    support : 7*5/total transactions  = 35/7500

```python
from apyori import apriori
sup = 7*5/len(df)
conf = 0.2
lift = 3

association_rules = apriori(records,min_support=sup,min_confidence=conf,min_lift=lift)
association_rules = list(association_rules)
association_rules[:10]
```

```python
results = []
for items in association_rules:
    #Extract product associations
    item_pair = items[0] #frozen set of items
    item_data = [i for i in item_pair] #save the items in a list

    support1 = items[1] # support 

    confidence1 = items[2][0][2]
    lift1 = items[2][0][3]

    results.append([item_data,round(support1,3),round(confidence1,3),round(lift1,2)])
print(results)
```

## Save these results inside a dataframe and share with the client

```python
df_results = pd.DataFrame(results)
df_results.head()
```

```python
df_results.columns = ["Product_Associations","Support","Confidence","Lift"]
df_results.head()
```

```python
df_results.to_csv("Product_Associations.csv",index=False)
```
