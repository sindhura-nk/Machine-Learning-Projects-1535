## Decision Tree Regressor

### Problem Statement: 
Business wants us to build a model that predicts the power energy generated by the powerplant on basis of the sensor readings

## Step1: Data Gathering

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
path=r"https://raw.githubusercontent.com/sindhura-nk/Datasets/refs/heads/main/PowerPlant.csv"
import pandas as pd
df = pd.read_csv(path)
df.head()
```

## Data Descriptions

    AT : ATMOSPHERIC TEMPERATURE
    AP: ATMOSPHERIC PRESSURE
    RH : RELATIVE HUMIDITY
    V : VACCUM
    PE : POWER ENERGY

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

## Separate X and Y features
Y  :Power energy

```python
X = df.drop(columns='PE')
Y = df[['PE']]
```

```python
X.head()
```

```python
Y.head()
```

## Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,train_size=0.80,random_state=21)
```

```python
print(xtrain.shape)
```

```python
print(xtest.shape)
```

## Data cleaning and data preprocessing

```python
X.columns
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler

con = list(X.columns)

con_pipe = make_pipeline(
    SimpleImputer(strategy='median'),
    MinMaxScaler()
).set_output(transform='pandas')
con_pipe
```

```python
con_pipe.fit(xtrain)
```

```python
xtrain_pre = con_pipe.transform(xtrain)
xtest_pre = con_pipe.transform(xtest)
```

```python
xtrain_pre.head()
```

```python
xtest_pre.head()
```

## Model Buidling

```python
from sklearn.tree import DecisionTreeRegressor
model= DecisionTreeRegressor(
    criterion='squared_error',
    max_depth = 8,
    min_samples_split = 4,
    min_samples_leaf=5
)
model.fit(xtrain_pre,ytrain)
```

```python
model.score(xtrain_pre,ytrain)
```

```python
model.score(xtest_pre,ytest)
```

## Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
params = {
    'criterion':['squared_error','absolute_error'],
    'max_depth':[2,3,4,5,6,7,8,9,10],
    'min_samples_leaf':[2,3,4,5,6,7],
    'min_samples_split':[2,4,6,8,10,12,14]
}

base_model = DecisionTreeRegressor()
gscv = GridSearchCV(estimator=base_model,param_grid=params,cv=3,scoring='r2')
gscv.fit(xtrain_pre,ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_dtr = gscv.best_estimator_
```

## MOdel evaluation

```python
best_dtr.score(xtrain_pre,ytrain)
```

```python
best_dtr.score(xtest_pre,ytest)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
ypreds = best_dtr.predict(xtest_pre)
ypreds = ypreds.round(2)
ypreds[:10]
```

```python
mse = mean_squared_error(ytest,ypreds)
rmse = mse**(1/2)
mae = mean_absolute_error(ytest,ypreds)
r2 = r2_score(ytest,ypreds)

print(f"MSE:{mse:.2f}")
print(f"RMSE:{rmse:.2f}")
print(f"MAE:{mae:.2f}")
print(f"r2:{r2*100:.2f}%")
```
