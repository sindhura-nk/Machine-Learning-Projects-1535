## Business Statement : Build a Regression model that predicts the profit of startups on basis of the investment they have made in RND (Research and Development)

## Step 1: Data Gathering

```python
path = r"https://raw.githubusercontent.com/sindhura-nk/Datasets/refs/heads/main/50_Startups.csv"
import pandas as pd
df = pd.read_csv(path)
df.head()
```

## Perform basic data quality checks

```python
df.shape
```

```python
df.info()
```

## EDA - Exploratory Data Analysis

## Seperate X and Y features

    X  :  RND
    Y  :  PROFIT

```python
X = df[['RND']]
Y = df[['PROFIT']]
```

```python
X.head()
```

```python
Y.head()
```

## Model Building

Profit_Predicted = B0 + B1.RND

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```

```python
model.coef_ # slope of the model
```

```python
model.intercept_ # y intercept
```

Profit_predicted = 49032.9 + (0.85)*RND

    If RND is 0, the profit predicted will be 49032
    If RND increases by 1000, you will observe that Profit predicted will increase by 850.

```python
X.head(10)
```

```python
ypreds = model.predict(X)
ypreds[:10] # these predictions are in array format
```

```python
Y.head(10)
```

## Evaluate the Model Performance 

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

mse = mean_squared_error(Y,ypreds)
mae = mean_absolute_error(Y,ypreds)
rmse = mse**(1/2)
r2 = r2_score(Y,ypreds)

print("----------------------")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"R squared error or R2 score: {r2*100:.2f}%")
print("----------------------")
```

## Above model is a good model as the r2 score is above 0.8. This model can be finalised for deployment
